### Reading NOCS Data

```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
df = spark.read.format("csv")\
            .option("header", True)\
            .option("inferSchema", True)\
            .load("abfss://bronze@olympics2024projectdl.dfs.core.windows.net/nocs")
```

```python
display(df)
```


**Dropping country column**

```python
df = df.drop("country")
```

```python
df = df.withColumn("tag", split("tag", "-")[0]) 
display(df)
```

```python
df.write.format("delta")\
        .mode("append")\
            .option("path", "abfss://silver@olympics2024projectdl.dfs.core.windows.net/nocs")\
                .saveAsTable("olympics.silver.nocs")
```

```python
df.write.format("delta")\
        .mode("append")\
        .saveAsTable("olympics.silver.nocs_man")
```
